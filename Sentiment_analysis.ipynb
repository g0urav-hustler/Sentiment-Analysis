```python
# importing the libraries
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.preprocessing import sequence
from keras.layers.embeddings import Embedding

import nltk
from nltk.tokenize import word_tokenize

import numpy as np
import pandas as pd
import re
```

```python
# loading the imdb dataset , top 5000 words
top_words = 5000
(x_train,y_train), (x_test,y_test) = imdb.load_data(num_words= top_words)
```

```python
# padding all sentences should be equal in length
max_review_length = 2000

x_train = sequence.pad_sequences(x_train, maxlen = max_review_length)
x_test = sequence.pad_sequences(x_test, maxlen = max_review_length)
```

```python
# Creating the model
vector = 32

model= Sequential()
model.add(Embedding(top_words, vector, input_length=max_review_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.2))
model.add(Dense(1, activation= 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
# fitting the model
model.fit(x_train, y_train, validation_data=(x_test,y_test), validation_split= 0.3,
          epochs = 30, batch_size = 64)
```

```python
accuracy = model.evaluate(x_test, y_test, verbose =0)
print("Accuracy =", accuracy[1]*100)
```
